In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

<h4 style="text-align: center;" markdown="1">  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Figure 1 from (Ferreira et al, 2014).</h4>

<br>

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

**Easiest option:** Implement the bar coloring as described above - a color scale with only three colors, (e.g. blue, white, and red). Assume the user provides the y axis value of interest as a parameter or variable.


**Harder option:** Implement the bar coloring as described in the paper, where the color of the bar is actually based on the amount of data covered (e.g. a gradient ranging from dark blue for the distribution being certainly below this y-axis, to white if the value is certainly contained, to dark red if the value is certainly not contained as the distribution is above the axis).

**Even Harder option:** Add interactivity to the above, which allows the user to click on the y axis to set the value of interest. The bar colors should change with respect to what value the user has selected.

**Hardest option:** Allow the user to interactively set a range of y values they are interested in, and recolor based on this (e.g. a y-axis band, see the paper for more details).

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])


df
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.widgets import Slider, Button, RadioButtons, Cursor, SpanSelector
from scipy.stats import sem, norm# standard error mean
from IPython.core.debugger import set_trace
%matplotlib notebook


class InteractiveBarChart(object):
    """Creates an interactive bar chart based on the dataframe passed.
    
    Keyword arguments:
    df -- DataFrame object to draw from
    axis -- Choices: 0 (columns), 1 (rows), 'auto' (will determine automatically,
            see the 'reshape_dataframe' method for more details).
    """
    df = None
    descriptions = None
    
    cursor = None
    
    fig = None
    ax = None
    
    config = None
    
    # color palette
    palette = None
    _std_palette = {
                'const_gradient': [
                    '#022F61',#0
                    '#1C64AE',#1
                    '#3F92C5',#2
                    '#90C5E0',#3
                    '#D1E5F1',#4
                    '#F7F7F7',#5
                    '#FEDBC6',#6
                    '#F6A57E',
                    '#D85F48',
                    '#B41427',
                    '#68001E',
                ],
                'range_gradient': [
                    '#FFF7EC',
                    '#FEE8C8',
                    '#FDD49E',
                    '#FDBB84',
                    '#FC8D59',
                    '#EF6548',
                    '#D7301F',
                    '#B30000',
                    '#7F0000',
                ],
                'background': '#000000',
                'widget_bg': '#f2e8bf',
                'widget_fg': '#83a9ea'                
            }
    
    def __init__(self, df, axis='auto', palette=None,
                 config='range'):
        self.config = config
        if axis == 'auto':
            self.df = self.reshape_dataframe(df)
        elif axis == 1:
            self.df = df.transpose()
        elif axis == 0:
            self.df = df
        else:
            raise ValueError('The axis value needs to be 0, 1 or \'auto\'.')
        if palette is not None:
            self.palette = palette
        else:
            self.palette = self._std_palette
        self.fig, self.ax = plt.subplots()
        
#         plt.subplots_adjust(left=0.25)
        
        self.describe_columns()
        
        self.redraw()
#         self.draw_radio_buttons()
        self.draw_annotation()
#         self.radio.on_clicked(self.change_selector)
        for i in range(len(self.rects)):
            self.rects[i].set_color('#c5c5c5')
            self.rects[i].set_ec('#403530')
            self.rects[i].set_linestyle('--')

    def show(self):
        return plt.show()

    
    def on_pick(self, event):
        X, Y = event.xdata, event.ydata
        
        self.recolor_constant(Y)
#     def draw_radio_buttons(self):
#         self.radio_ax = plt.axes([0.025, 0.5, 0.15, 0.15],
#                                  facecolor=self.palette['widget_bg'])
#         self.radio = RadioButtons(self.radio_ax, ('constant', 'range'),
#                              active=0, activecolor=self.palette['widget_fg'])
        

    def describe_columns(self):
        self.descriptions = self.df.apply(lambda x: x.describe(), axis=0)
        sums = self.df.apply(lambda x: x.sum(), axis=0)
        stderrs = df.apply(lambda x: sem(x) * 1.96, axis=1)
        
        self.descriptions.loc['sum'] = sums
        self.descriptions.loc['stderr'] = stderrs

    def reshape_dataframe(self, df):
        """Will reshape the dataframe based on its shape, preferring
        a longer dataframe shape than a wider one; will transpose self.df
        in case of a wider one.
        """
        rows, cols = df.shape
        if rows > cols:
            return df
        else:
            return df.transpose()

    def draw_annotation(self):
        trans = self.ax.get_xaxis_transform() # x in data untis, y in axes fraction
        self.annot = self.ax.annotate('', xy=(-0.6, 0.85 ), xycoords=trans,
                                 bbox=dict(boxstyle='round', fc=self.palette['widget_bg']))
        self.annot.set_visible(False)
        self.fig.canvas.draw_idle()

    def redraw(self):
        ind = np.arange(len(self.df.columns))
        self.ax.cla()
        self.rects = self.ax.bar(ind, self.descriptions.loc['mean'], 0.95, color='#c5c5c5',
                                 yerr=self.descriptions.loc['stderr'], capsize=3)
        for rect in self.rects:
            rect.set_ec('#403530')
            rect.set_linestyle('--')
        self.ax.set_xticks(ind)
        self.ax.set_xticklabels(self.df.columns)
        
        if self.config == 'range':
            self.draw_span_selector()
        elif self.config == 'constant':
            self.draw_cursor()
            self.cid = self.fig.canvas.mpl_connect('button_press_event', self.on_pick)
        self.draw_annotation()
        self.fig.canvas.draw_idle()
    
    def update_annotation(self, vmin, vmax=None):
        if self.config == 'range':
            text = 'Selection:\n Max: {}\n Min: {}'.format(
                round(vmax,2),
                round(vmin,2))
        elif self.config == 'constant':
            text = 'Selection:\n Val: {}'.format(
                round(vmin,2))
        self.annot.set_text(text)
        self.annot.set_visible(True)
        self.fig.canvas.draw_idle()

        
    def recolor_range(self, vmin, vmax):
        """Takes 100 random samples from the selected range, and tests them against
        the normal distribution for each rect (with mean [mu] and std [sigma] as described in
        self.descriptions) and tests whether they are likely to be inside this distribution."""
        self.redraw()
        samples = np.random.choice(np.arange(vmin, vmax), 100)
        print(samples)
        # gradient value distribution
        vals = [
            (0, 0.11, 0),
            (0.11, 0.22, 1),
            (0.22, 0.33, 2),
            (0.33, 0.44, 3),
            (0.44, 0.56, 4),
            (0.56, 0.67, 5),
            (0.67, 0.78, 6),
            (0.78, 0.89, 7),
            (0.89, 1, 8),
        ]
        for i in range(len(self.rects)):
            mu = self.rects[i].get_height()
            sigma = self.descriptions.loc['stderr'][self.df.columns[i]]
            val = norm.cdf(samples, loc=mu, scale=sigma).mean()
            for v in vals:
                if val >= v[0] and val < v[1]:
                    self.rects[i].set_color(self.palette['range_gradient'][v[2]])
        self.update_annotation(vmin=vmin, vmax=vmax)
        self.fig.canvas.draw_idle()
        

    def recolor_constant(self, y_value):
        self.redraw()
        # gradient value distribution
        vals = [
            (0, 0.09, 0),
            (0.09, 0.18, 1),
            (0.27, 0.36, 2),
            (0.36, 0.45, 3),
            (0.45, 0.55, 4),
            (0.55, 0.64, 5),
            (0.64, 0.73, 6),
            (0.73, 0.82, 7),
            (0.82, 0.91, 8),
            (0.91, 1, 9),
        ]
        for i in range(len(self.rects)):
            mu = self.rects[i].get_height()
            sigma = self.descriptions.loc['stderr'][self.df.columns[i]]
            val = norm.cdf(y_value, loc=mu, scale=sigma)
            for v in vals:
                if val >= v[0] and val < v[1]:
                    self.rects[i].set_color(self.palette['const_gradient'][v[2]])
        plt.plot([0, len(self.df.columns) - 1], [y_value, y_value], color='orange', linestyle='-', linewidth=1)
        self.update_annotation(vmin=y_value)
        self.fig.canvas.draw_idle()
        

    def draw_cursor(self):
        self.cursor = Cursor(self.ax,
                                 color=self.palette['widget_fg'], lw=1,
                                 horizOn=True, vertOn=False)
        
    def draw_span_selector(self):
        self.span = SpanSelector(self.ax, self.recolor_range, 'vertical',
                                 rectprops=dict(facecolor=self.palette['widget_fg'],
                                                alpha=0.5), span_stays=True)

```

```python

const_plot = InteractiveBarChart(df, config='constant')
```
